```python
from colored_networks.colored_network import *
```

```python
n = 3
m = 5
```

```python
n_nodes = [Node(i) for i in range(n+1)]
m_nodes = [Node(i) for i in range(n+1, n+1+m+1)]
edges = ([Edge(n_nodes[-1], m_nodes[0], "red")]
    + [Edge(i, j, "blue") for i, j in zip(n_nodes[:-1], n_nodes[1:])]
    + [Edge(i, j, "blue") for i, j in zip(m_nodes[:-1], m_nodes[1:])])
```

```python
rules = [
    DeleteRule(
        name="Add numbers",
        input={"blue": 1, "red": 1},
        rewiring={("blue", "red"): "blue"},
    ),
    DeleteRule(
        name="Add 0",
        input={"red": 1},
        rewiring={},
    ),
]
```

```python
net = ColoredNetwork(n_nodes + m_nodes, edges, rules)
net.reduce(verbose=True)
```
